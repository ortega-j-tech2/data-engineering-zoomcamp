```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
parquet_name = 'taxi_data_parquet.parquet'
csv_name = 'output.csv'
```

```python
df_parquet = pd.read_parquet(parquet_name)
print('Converting file...', flush=True)
df_parquet.to_csv(csv_name)
print('Converting file finished!', flush=True)
```

```python
print('reading file for conversion...', flush=True)
df_parquet = pd.read_csv(csv_name)
print('reading file for conversion...done!', flush=True)
# If you know the name of the column skip this
first_column = df_parquet.columns[0]
# Delete first
print('dropping column from df...', flush=True)
df_parquet = df_parquet.drop([first_column], axis=1)
print('dropping column from df...done!', flush=True)
print('saving file...', flush=True)
df_parquet.to_csv(csv_name, index=False)
print('saving file...done!',flush=True)
```
